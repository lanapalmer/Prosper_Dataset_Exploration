## Dataset Overview

> Prosper Marketplace, Inc. is a privately-held, San Francisco-based company in the peer-to-peer lending industry. This dataset includes information on loans funded through the marketplace.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
dateparse = lambda x: pd.datetime.strptime(x, '%Y-%m-%d %H:%M:%S')
#Import Prosper csv
prosper = pd.read_csv('prosperLoanData.csv', parse_dates=['ListingCreationDate'], date_parser = pd.to_datetime)
```

```python
# Copy Data
prosper_c = prosper.copy()

#Cleaning data
#Drop rows with null ProsperScore values
prosper_c.dropna(subset=['ProsperScore'], inplace=True)
#Drop scores over 10.0
prosper_c.drop(prosper_c[prosper_c['ProsperScore'] > 10.0].index, inplace = True) 
#Filtering only dates beginning on July 1, 2009
prosper_c = prosper_c[(prosper_c['ListingCreationDate'] >= '2009-07-01')]

```

## The ProsperScore: Average Score

The ProsperScore is a custom risk score built using historical Prosper data. Each borrower is assigned a score ranging from 1.0 (highest risk) to 10.0 (lowest risk). Lenders may use this score when choosing loans to fund. The next slide shows the range of ProsperScores, with an average of 5.86. 


```python
#Plot Boxplot
plt.figure(figsize=(14, 10))
sns.set_style('whitegrid')
ax = sns.boxplot(data=[prosper_c['ProsperScore']])
ax.set_title('Boxplot of ProsperScore');
```

## The ProsperScore: Score Range

> As the next slide illustrates, most borrowers have scores between 4 and 8, and very few have scores below 2.

```python

#Plot Histogram
plt.figure(figsize=(60, 60))
prosper_c.hist(column='ProsperScore')
plt.ylabel('Frequency', fontsize=16)
plt.xlabel('ProsperScore', fontsize=16)
plt.title('Histogram of ProsperScore', fontsize=16);
```

## ProsperScore vs Lender Yield

The Lender Yield is equal to the interest rate on the loan less the Prosper servicing fee. As we see in the following slide, the lowest ranked borrowers, with a ProsperScore of 1.0, have a mean Lender Yield which is dramatically higher than the highest-ranked borrowers.

```python
#Plot Violinplot
plt.figure(figsize=(16, 10))

sns.violinplot(data = prosper_c, x = 'ProsperScore', y = 'LenderYield', color = 'blue',
              inner = None)
plt.title('Violinplot of ProsperScore vs Lender Yield', fontsize=16)

plt.ylabel('Lender Yield', fontsize=16)
plt.xlabel('ProsperScore', fontsize=16);
```

## ProsperScore, Lender Yields and Investors

The number of investors for each loan in the dataset ranged from 1 to 1200, with an average of 80 investors per loan. The next slide shows the correlation between the number of investors, prosper score, and lender yield. We see that:

- LenderYield and ProsperScore have a strong negative correlation (-0.64)
<br>
<br>
- LenderYield and the number of Investors have a negative correction (-0.24)
<br>
<br>
- The ProsperScore and the number of investors have a positive correlation (0.33)


```python
prosper_s = prosper_c[['LenderYield', 'ProsperScore', 'Investors']]
```

```python
#Plotting Heatmap of LenderYield, ProsperScore, and Investors
plt.figure(figsize=(16, 10))
sns.set(font_scale=2)
plt.title('Heatmap of LenderYield, ProsperScore, and Investors', fontsize=16)
sns.heatmap(prosper_s.corr(), annot = True, fmt = '.2f', cmap = 'vlag_r', center = 0);
```

This preliminary datset exploration has revealed investors who are willing to take the risk on borrowers with lower ProsperScores and fewer lenders are rewarded by higher returns. However, these borrowers are a rarity in the marketplace, with most borrowers falling in the 4.0-6.0 score range.
